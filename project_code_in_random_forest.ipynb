```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import numpy as np

# Load the dataset
data = pd.read_csv('complete soil data.csv')

print("Dataset loaded successfully")

# Display the first few rows of the dataset
print(data.head())

# Define features and target variable
feature_columns = ['PH', 'EC', 'OC', 'N', 'P2O5', 'K20', 'S', 'CU', 'FC', 'MN', 'ZN', 'BA', 'Temparature', 'Humidity', 'Rainfall']
features = data[feature_columns]
target = data['CROP']

# --- Add data cleaning steps here ---
# Convert feature columns to numeric, coercing errors
for col in feature_columns:
    # Convert column to numeric, replacing errors with NaN
    data[col] = pd.to_numeric(data[col], errors='coerce')

# Check for and display rows with non-numeric values (now NaN) in features
nan_rows = data[data[feature_columns].isnull().any(axis=1)]
if not nan_rows.empty:
    print("\nRows with non-numeric values in feature columns:")
    print(nan_rows)
    print(f"\nFound {len(nan_rows)} rows with non-numeric values. Dropping these rows...")
    # Drop rows where any of the feature columns contain NaN
    data.dropna(subset=feature_columns, inplace=True)
else:
    print("\nNo non-numeric values found in feature columns.")

# Re-define features and target variable after cleaning
features = data[feature_columns]
target = data['CROP']
# --- End of data cleaning steps ---


# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Create a Random Forest Classifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
# This should now work after removing rows with non-numeric values
rf.fit(X_train, y_train)

# Make predictions
y_pred = rf.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')

# Generate classification report
report = classification_report(y_test, y_pred)
print("Classification Report:\n", report)

# Feature importance
importances = rf.feature_importances_
feature_importance = pd.DataFrame(importances, index=features.columns, columns=['Importance']).sort_values('Importance', ascending=False)


# Predict recommended crop for a given soil condition
# Ensure sample_data is a DataFrame or numpy array with the correct structure
sample_data = [[6.5, 0.2, 0.5, 120, 50, 80, 15, 0.3, 25, 30, 0.2, 40, 28, 65, 180]]  # Replace with actual soil values
sample_data_df = pd.DataFrame(sample_data, columns=feature_columns) # Convert to DataFrame with correct column names
recommended_crop = rf.predict(sample_data_df) # Predict using the DataFrame
print("\nRecommended Crop:", recommended_crop[0])
```
